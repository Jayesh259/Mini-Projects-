## 6. Preprocess Historical Data

```python
hist_df['Open Time'] = pd.to_datetime(hist_df['Open Time']/1000 , unit ='s')
```

```python
hist_df['Close Time'] = pd.to_datetime(hist_df['Close Time']/1000 , unit ='s')
```

```python
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Quote Asset Volume', 'TB Base Volume', 'TB Quote Volume']
```

```python
hist_df[numeric_cols] = hist_df[numeric_cols].apply(pd.to_numeric, axis = 1)
```
